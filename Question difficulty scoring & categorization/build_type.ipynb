```python
# This file is to get ordered categorization of each question in testbench.
# (When we run the pipeline of QTSA generation, questions are generated in order)
```

```python
import json

input_file = "SFT/Qwen3-0.6B/testbench/mcQTSA_test_final.jsonl"
output_file = "testbench_type.jsonl"

# id % 5 
type_mapping = {
    1: "Conceptual Understanding",
    2: "Methematical Derivation",
    3: "Design Implementation",
    4: "Performance Analysis",
    0: "Application & Limitations"
}

with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", encoding="utf-8") as fout:

    for line in fin:
        data = json.loads(line)

        idx = int(data["id"])

        data["type"] = type_mapping[idx % 5]

        fout.write(json.dumps(data, ensure_ascii=False) + "\n")

print(f"Output file: {output_file}")
```
